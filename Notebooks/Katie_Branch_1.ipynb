```python
import requests
import matplotlib.pyplot as plt
import json
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
```

```python
endpoint = "https://data.nashville.gov/resource/2u6v-ujjs.json"
```

```python
type(endpoint)
```

```python
params={'offense_description': 'BURGLARY- AGGRAVATED', '$where': "incident_reported between '2022-01-01' and '2022-09-30'", '$limit': 50000}
response = requests.get(endpoint, params=params)
response
```

```python
response.text
```

```python
response
```

```python
res = response.json()
res
```

```python
burglaries = pd.DataFrame(res)
burglaries
```

```python
burglaries.info()
```

```python
burglaries["latitude"] = pd.to_numeric(burglaries["latitude"])
burglaries["longitude"] = pd.to_numeric(burglaries["longitude"])
```

```python
burglaries.info()
```

```python
burglaries["geometry"] = burglaries.apply(lambda x: Point((x['longitude'], x['latitude'])), axis=1)
```

```python
burglaries["geometry"] = gpd.points_from_xy(burglaries["longitude"], burglaries["latitude"])
```

```python
burglaries
```

```python
burglaries.head()
```

```python
census_tracts = gpd.read_file("../data/tl_2020_47_tract.shp")
census_tracts
```

```python
burglaries = gpd.GeoDataFrame(burglaries, geometry=burglaries["geometry"], crs=census_tracts.crs)
burglaries
```
